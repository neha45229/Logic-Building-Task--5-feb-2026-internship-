1.Social Media – Post Engagement Analyzer


```python
def calculate_viral_post(likes):
  total_likes = sum(likes)
  print(f"Total Likes:",total_likes)
  if total_likes >= 1000:
    print("Post Status:Viral Post")
  else:
    print("Post Status:Normal Engagement")
likes = [250,390,426,480,578,600]
calculate_viral_post(likes)

```

2.Healthcare – Medicine Stock Alert System


```python
def stock_availability(Stock):
  for i in range(Stock):
    if Stock < 10 :
      print("Status:Low Stock Alert")
      break
    elif Stock >= 10:
      print("Status:Stock Sufficient")
      break
    else:
      print("Status:Invalid Stock")
      break
Stock = int(input("Medicine Stock:"))
stock_availability(Stock)
```

3. Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall):
  average=int(sum(rainfall)/len(rainfall))
  print("Average Rainfall :",average)
  required_level=12
  if average >= required_level:
    print("Rainfall Status:Adequate Rainfall")
  else:
    print("Rainfall Status:Inadequate Rainfall")
rainfall=[42,64,54,27,158,87]
rainfall_checker(rainfall)
```

4.Social Media – Duplicate Account Detection


```python
def account_detection(accounts):
  found=False
  for i in range(len(accounts)):
    for j in range(i+1,len(accounts)):
      if accounts[i] == accounts[j]:
        found=True
        break
    if found:
      print("Duplicate Accounts Found: Yes")
      break
    else:
      print(" Accounts are unique")
accounts=["raj","sunny","pari","raj","mohan","parul","pari"]
account_detection(accounts)
```

7.System Monitoring – Application Health

```python
def healthy_patients(Errors):
  Error_count=0
  for i in range(Errors):
     if Errors == 0:
       print("System Status:Healthy")
       break
     elif Errors <= 5:
       print("System Status:Minor Issues")
       break
     elif Errors >= 5:
       print("System Status:Critical Issues")
       break
     else:
       print("Invalid Application")
       break
     Error_count+=1
Errors=int(input("Error count:"))
healthy_patients(Errors)

```

9.E-Learning – Student Attendance Eligibility

10.Smart Electricity Bill Analyzer


```python
def electricity_analyzer(bill):
  total_units=sum(bill)
  print("Total Electricity units:",total_units)
  if total_units ==100 :
    amount=total_units*3
  elif total_units<=200:
    amount=total_units*5
  else:
    amount=total_units*7
  print("Total bill amount:",amount)
  if amount< 500:
    print("Electricity Status:Low usage")
  elif amount<=1500:
    print("Electricity Status:Moderate Usage")
  else:
    print("Electricity Status:High Usage")
bill=[78,45,67,21]
electricity_analyzer(bill)


```
